# check for file presence

```python

import os
def list_folders(directory):
    # Get a list of all entries in the directory
    entries = os.listdir(directory)

    # Filter out non-directory entries
    folders = [
        os.path.join(directory, entry) 
        for entry in entries 
        if os.path.isdir(os.path.join(directory, entry))
    ]
    return folders

def does_file_exist(directory, filename):
    """
    Check if a specific file exists in a directory.

    Parameters:
    - directory (str): The path to the directory.
    - filename (str): The name of the file to check.

    Returns:
    - bool: True if the file exists, False otherwise.
    """
    file_path = os.path.join(directory, filename)
    return os.path.exists(file_path) and os.path.isfile(file_path)
#symbol_folders = list_folders(path)
#symbol_file = [does_file_exist(x,filename) for x in symbol_folders]
#print(len(symbol_folders),sum(symbol_file))
```

# check for changes ratio within a day

```python
import pandas as pd 

symbol_folders = list_folders("mocks3yfinance/daily")


filename = '2023-12-25.parquet'

symbol_file = {
    symbol:os.path.join(path,filename)
    for symbol,path in [
        (x.split('/')[-1],x) for x in symbol_folders
    ]
}

dfs = {}
for k in sorted(symbol_file.keys()):
    print(f"reading {k}")
    dfs[k] = pd.read_parquet(symbol_file[k])

max_possible_gain_ratio = []
close_change_ratio = []
for k,df in dfs.items():
    print(f'doing calc on {k}')
    max_possible_gain_ratio += ((df['High'] - df['Open'])/df['Open'] ).tolist()
    close_change_ratio += ((df['Close'] - df['Open'])/df['Open'] ).tolist()
```

```python
c = close_change_ratio
print(len(c))
# mean
print(sum(c)/len(c))
# max
print(max(c))

```

```python
h = pd.Series(max_possible_gain_ratio)
print(h.median())
print(h.mean())
print(h.max())

```
